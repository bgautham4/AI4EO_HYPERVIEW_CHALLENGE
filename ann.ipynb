```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_fn(x):
    return x
```

```python
#linear activation
x = np.linspace(-5,5,100)
y = linear_fn(x)
plt.plot(x,y)
plt.grid()
```

```python
#hard limit activation
y2 = np.heaviside(x,1)
plt.plot(x,y2)
plt.grid()
```

```python
#tanh activation
y3 = np.tanh(x)
plt.plot(x,y3)
plt.grid()
```

```python
def sigmoid(x,a):
    return 1/(1+np.exp(-a*x))
```

```python
#sigmoid activation
y4 = sigmoid(x,1)
plt.plot(x,y4)
plt.grid()
```
